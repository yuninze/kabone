```python
import os
import pandas as pd
import numpy as np
```

```python
trSource_=pd.read_clipboard()
```

```python
trSource_.info()
```

```python
x=trSource_.melt(["idx","name","klass","sex"],trSource.columns[4:])
x.sample(3).iloc[:,2:]
```

```python
x.loc[:,"variable"]=x.variable.apply(lambda q:q.split(","))
x.sample(3).iloc[:,2:]
```

```python
listedValues=np.column_stack([x.variable.tolist()])
listedValues
```

```python
listedValuesToColumns=pd.DataFrame(listedValues,columns=["trainingPeriod","trainingWeek"])
listedValuesToColumns
```

```python
trainingSource=pd.concat([x,listedValuesToColumns],axis=1)
trainingSource.sample(3).iloc[:,2:]
```

```python
trainingSource.loc[:,"value"]=trainingSource.loc[:,"value"].fillna("#")

trainingClass=trainingSource["value"].apply(
	lambda q:q[:q.find("-")]
)

trainingCompany=trainingSource["value"].apply(
	lambda q:q[q.find("-")+1:]
).apply(
	lambda q:list(q)
).apply(
	lambda q:[w for w in q if not w.isdigit()]
).apply(
	lambda q:"".join([w for w in q if not w=="-"])
)

trainingSource["trainingClass"]=trainingClass.copy()
trainingSource["trainingCompany"]=trainingCompany.copy()
```
